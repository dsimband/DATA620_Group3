## Libaries

The first step to completing this assignment is to import the necessary Python libraries. NetworkX will be used to create the graph and add the necessary edges to the graph. Matplotlib will be used to display the graph. 

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Initialize Graph Object and Add Edges

In the code chunk below, I initialized an empty graph object and populated the graph with edges. Each edge corresponds to a relationship between two people. 

```python
people = nx.Graph() 

# Carol
people.add_edge("Carol", "Andre")
people.add_edge("Carol", "Diane")
people.add_edge("Carol", "Fernando")

# Andre
people.add_edge("Andre", "Diane")
people.add_edge("Andre", "Beverly")
people.add_edge("Andre", "Fernando")

# Fernando
people.add_edge("Fernando", "Diane")
people.add_edge("Fernando", "Garth")
people.add_edge("Fernando", "Heather")

# Garth
people.add_edge("Garth", "Diane")
people.add_edge("Garth", "Beverly")
people.add_edge("Garth", "Ed")
people.add_edge("Garth", "Heather")

# Ed
people.add_edge("Ed", "Diane")
people.add_edge("Ed", "Beverly")

# Beverly
people.add_edge("Beverly", "Diane")

# Heather
people.add_edge("Heather", "Ike")

# Ike
people.add_edge("Ike","Jane")
```

## Draw and Display Graph

The nx.draw function uses the information collected in the 'people' graph object to draw a graph. The plt.show function is used to display the graph.

```python
# Draw Graph
nx.draw(people, with_labels = True)

# Display Graph
plt.show()
```
